# Introduction to Hugging Face 🤗

Welcome to this beginner-friendly guide to **Hugging Face Transformers**! 

**What is Hugging Face?**
Hugging Face is like the 'GitHub for Machine Learning'. It hosts hundreds of thousands of pre-trained Artificial Intelligence models that you can download and use for free. Instead of spending millions of dollars and months of time training a model from scratch, you can download a state-of-the-art model in just a few lines of code.

### 💡 Pro-Tip: How to Use AI Without Wasting Compute/Tokens
When you are a beginner, it's easy to accidentally waste compute resources (like Colab GPU limits or API tokens). Here is how you avoid that:
1. **Use small models for testing:** Instead of downloading a massive 10-Billion parameter model, test your code with small models (like `distilbert`) first.
2. **Hugging Face caches models locally:** The first time you run `pipeline()`, it downloads the model to your hard drive. If you run the cell again, it will load it instantly from your hard drive instead of re-downloading it. Don't delete the cache!
3. **Use Pipelines:** The `pipeline` function is the easiest way to use AI. It handles all the complex preprocessing (tokenization) and postprocessing for you, so you don't make mistakes that waste compute.

# Tokenization

## 🔢 Tokenization: How AI Reads
Computers cannot understand English text; they only understand numbers. 
A **Tokenizer** breaks down your sentence into small chunks (called 'tokens'—which can be words or parts of words) and converts them into numeric IDs that the neural network can process.

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, DistilBertTokenizer, DistilBertForSequenceClassification
```

```python
model_name2 = "nlptown/bert-base-multilingual-uncased-sentiment"
mymodel2 = AutoModelForSequenceClassification.from_pretrained(model_name2)
mytokenizer2 = AutoTokenizer.from_pretrained(model_name2)

classifier = pipeline("sentiment-analysis", model = mymodel2 , tokenizer = mytokenizer2)
res = classifier("I was so not happy with the Barbie Movie")
print(res)
```

```python
from transformers import AutoTokenizer

# Load a pre-trained tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Example text
text = "I was so not happy with the Barbie Movie"

# Tokenize the text
tokens = tokenizer.tokenize(text)
print("Tokens:", tokens)
```

```python
# Convert tokens to input IDs
input_ids = tokenizer.convert_tokens_to_ids(tokens)
print("Input IDs:", input_ids)
```

```python
# Calling the tokenizer directly on text does BOTH tokenization and converting to IDs.
# It also adds special tokens like [CLS] (start) and [SEP] (separator).
encoded_input = tokenizer(text)
print("Encoded Input:", encoded_input)
```

```python

# Decode the text
decoded_output = tokenizer.decode(input_ids)
print("Decode Output: ", decoded_output)
```

**token_type_ids**<br>
These IDs are used to distinguish between different sequences in tasks that involve multiple sentences, such as question-answering and sentence-pair classification. BERT uses this mechanism to understand which tokens belong to which segment. For single-sequence tasks like sentiment analysis, token_type_ids are all zeros.

**attention_mask** <br>
The attention mask is used to differentiate between actual tokens and padding tokens (if any). It helps the model focus on non-padding tokens and ignore padding tokens. A value of 1 indicates that the token should be attended to, while a value of 0 indicates padding.

**Why Padding Tokens Are Used**<br>
Uniform Sequence Length: Deep learning models typically process input data in batches. To efficiently process these batches, all sequences in a batch must have the same length. Padding tokens ensure this by extending shorter sequences to match the length of the longest sequence in the batch.
Efficient Computation: Fixed-length sequences allow for more efficient use of hardware resources, as the model can process all sequences in parallel without needing to handle variable-length sequences individually.



# Fine Tunning IMDB

## Step 1: Install Necessary Libraries

```python
!pip install datasets
```

## Step 2: Load and Prepare the Dataset

```python
from datasets import load_dataset
dataset = load_dataset('imdb')
```

```python
dataset
```

```python
dataset["train"][0]
```

## Step 3: Preprocess the Data
Tokenize the dataset using the tokenizer associated with the pre-trained model.

```python
from transformers import AutoTokenizer

# Load the tokenizer
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples['text'], padding="max_length", truncation=True)

tokenized_datasets = dataset.map(tokenize_function, batched=True)
```

```python
tokenized_datasets
```

```python
tokenized_datasets["train"][0]
```

## Step 4: Set Up the Training Arguments
Specify the hyperparameters and training settings.

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir='./results',          # Output directory
    eval_strategy ="epoch",     # Evaluate every epoch
    learning_rate=2e-5,              # Learning rate
    per_device_train_batch_size=16,  # Batch size for training
    per_device_eval_batch_size=16,   # Batch size for evaluation
    num_train_epochs=1,              # Number of training epochs
    weight_decay=0.01,               # Strength of weight decay
)
training_args
```

## Step 5: Initialize the Model
Load the pre-trained model and define the training procedure.

```python
from transformers import AutoModelForSequenceClassification, Trainer

# Load the pre-trained model
model = AutoModelForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)

# Initialize the Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_datasets['train'],
    eval_dataset=tokenized_datasets['test']
)
```

## Step 6: Train the Model
Fine-tune the pre-trained model on your specific dataset.

```python
# Train the model
trainer.train()
```

## Step 7: Evaluate the Model
Assess the model's performance on a validation set.

```python
# Evaluate the model
results = trainer.evaluate()
print(results)
```

## Step 8: Save the Fine-Tuned Model
Save the fine-tuned model for later use.

```python
# Save the model
model.save_pretrained('./fine-tuned-model')
tokenizer.save_pretrained('./fine-tuned-tokenizer')
```

## 🎓 Mini-Project: Academic Paper Summarizer
In this mini-project, we will build a practical AI tool! Reading academic papers is time-consuming because the abstracts are very dense and technical. 

**What this project does:**
1. It uses the `arxiv` API to automatically search the internet and download the latest research papers about Artificial Intelligence.
2. It extracts the long, complex abstract from the paper.
3. It passes that abstract through a Hugging Face text summarization model (`facebook/bart-large-cnn`).

**Expected Output:** 
The model will read a giant paragraph of technical jargon and spit out a simple 2-3 sentence summary that anyone can understand! This is a real-world use case for Generative AI.

```python
!pip install arxiv
```

```python
import arxiv
import pandas as pd
```

```python
# Query to fetch AI-related papers
query = 'ai OR artificial intelligence OR machine learning'
search = arxiv.Search(query=query, max_results=10, sort_by=arxiv.SortCriterion.SubmittedDate)

# Fetch papers
papers = []
for result in search.results():
    papers.append({
      'published': result.published,
        'title': result.title,
        'abstract': result.summary,
        'categories': result.categories
    })

# Convert to DataFrame
df = pd.DataFrame(papers)

pd.set_option('display.max_colwidth', None)
df.head(10)
```

```python
# 1. Grab the long, complex abstract from the first paper we downloaded
abstract = df['abstract'][0]

# 2. Load the 'facebook/bart-large-cnn' model which is specifically trained to summarize news and articles
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# 3. Run the summarization pipeline on our abstract!
summarization_result = summarizer(abstract)
```
